```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
df = pd.read_csv("ACS_16_5YR_B27001_with_ann.csv",
                usecols = ['GEO.id','GEO.id2','GEO.display-label',
                           'HD01_VD11','HD01_VD14','HD01_VD17','HD01_VD20','HD01_VD23',
                            'HD01_VD26','HD01_VD29','HD01_VD39','HD01_VD42',
                           'HD01_VD45','HD01_VD48','HD01_VD51','HD01_VD54','HD02_VD57'])
```

```python
df = df[1:]
```

```python
df[['HD01_VD11','HD01_VD14','HD01_VD17','HD01_VD20','HD01_VD23',
                            'HD01_VD26','HD01_VD29','HD01_VD39','HD01_VD42',
                           'HD01_VD45','HD01_VD48','HD01_VD51','HD01_VD54','HD02_VD57']] = df[['HD01_VD11','HD01_VD14','HD01_VD17','HD01_VD20','HD01_VD23',
                            'HD01_VD26','HD01_VD29','HD01_VD39','HD01_VD42',
                           'HD01_VD45','HD01_VD48','HD01_VD51','HD01_VD54','HD02_VD57']].astype(float)
```

```python
df.dtypes
```

```python
df['uninsured adult male'] = df[[ 'HD01_VD11','HD01_VD14','HD01_VD17','HD01_VD20','HD01_VD23',
                            'HD01_VD26','HD01_VD29']].sum(axis=1)
```

```python
df['uninsured adult female'] = df[['HD01_VD39','HD01_VD42',
                           'HD01_VD45','HD01_VD48','HD01_VD51','HD01_VD54','HD02_VD57']].sum(axis=1)
```

```python
df_uninsured = df[['GEO.id','GEO.id2','GEO.display-label','uninsured adult male','uninsured adult female']]
```

```python
df_uninsured['uninsured total'] = df_uninsured['uninsured adult male'] + df_uninsured['uninsured adult female']
```

```python
df_uninsured
```

```python
df_pop = pd.read_csv("TEST-census-data-for-edit-test.csv")
```

```python
df_pop.head()
```

```python
df_pop = df_pop[['GEO.id','GEO.id2','GEO.display-label','HC01_VC03','HC01_VC33','HC01_VC34']]
```

```python
df_pop = df_pop[1:]
```

```python
df_pop[['HC01_VC03','HC01_VC33','HC01_VC34']] = df_pop[['HC01_VC03','HC01_VC33','HC01_VC34']].astype(float)
```

```python
df_pop
```

```python
df_pop['adult male'] = round(df_pop['HC01_VC03'] * df_pop['HC01_VC33'] / 100)
```

```python
df_pop['adult female'] = round(df_pop['HC01_VC03'] * df_pop['HC01_VC34'] / 100)
```

```python
df_pop['total'] = df_pop['adult male'] + df_pop['adult female']
```

```python
df_merge = pd.merge(df_uninsured,df_pop, on='GEO.id')
```

```python
df_merge = df_merge.drop(['HC01_VC03','HC01_VC33','HC01_VC34'], axis=1) 
```

```python
df_merge['uninsured rate'] = round(df_merge['uninsured total'] / df_merge['total'] * 100)
```

```python
df_merge['male uninsured rate'] = round(df_merge['uninsured adult male'] / df_merge['adult male'] * 100)
```

```python
df_merge['female uninsured rate'] = round(df_merge['uninsured adult female'] / df_merge['adult female'] * 100)
```

```python
df_merge['gender gap'] = df_merge['male uninsured rate'] - df_merge['female uninsured rate']
```

```python
df_merge = df_merge.sort_values(by='uninsured rate',ascending=False)
```

```python
df_merge.to_csv("uninsured adults.csv", index=False)
```
