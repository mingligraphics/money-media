```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
df_voting = pd.read_csv('TEST-census-data-for-edit-test.csv',
                  usecols = ['GEO.id','GEO.id2','GEO.display-label','HC01_VC03','HC02_VC03','HC01_VC108','HC02_VC108','HC01_VC109','HC01_VC110'])
```

```python
df_voting = df_voting[1:]
```

```python
df_voting.head()
```

```python
df_voting[['HC01_VC03','HC02_VC03','HC01_VC108','HC02_VC108','HC01_VC109','HC01_VC110']] = df_voting[['HC01_VC03','HC02_VC03','HC01_VC108','HC02_VC108','HC01_VC109','HC01_VC110']].astype(float)
```

```python
df_voting['2012 - 2016 voting ratio'] = df_voting.HC02_VC108 / df_voting.HC02_VC03 
df_voting['2007 - 2011 voting ratio'] = df_voting.HC01_VC108 / df_voting.HC01_VC03
```

```python
df_voting['increase?'] = df_voting['2012 - 2016 voting ratio'] > df_voting['2007 - 2011 voting ratio']
```

```python
df_voting['increase'] = df_voting['2012 - 2016 voting ratio'] - df_voting['2007 - 2011 voting ratio']
```

```python
df_increase = df_voting[df_voting['increase?'] == True]
```

```python
df_red = pd.read_csv("red_state.csv")
```

```python
df_merge = pd.merge(df_increase,df_red,left_on='GEO.display-label',right_on='state')
```
